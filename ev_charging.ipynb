```python
df["Charging_Hour"]=df["Charging_Start_Time"].dt.hour
```

```python
df["Month"]=df["Charging_Start_Time"].dt.month
```

```python
df["Month_Name"]=df["Charging_Start_Time"].dt.month_name()
```

```python
df["Revenue_Per_Kwh"]=df["Charging_Cost_INR"]/df["Energy_Consumed_kWh"]
```

```python
df["Duration_Category"]=pd.cut(
    df["Charging_Duration_Minutes"],
    bins=[0,45,90,200],
    labels=["Short","Medium","Long"]
)
```
